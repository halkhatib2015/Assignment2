Task 9. Display total number of players for each foot type according to their positions. But the result of each position should be a separate graph yet shown collectively 

```python
df_M = dffifa[dffifa['position'] == 'M']
df_A = dffifa[dffifa['position'] == 'A']
df_D = dffifa[dffifa['position'] == 'D']
df_GK = dffifa[dffifa['position'] == 'GK']
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 8))

sns.countplot(x='foot', data=df_GK, ax=axes[0, 0])
axes[0, 0].set_title('Goalkeeper')

sns.countplot(x='foot', data=df_M, ax=axes[0, 1])
axes[0, 1].set_title('Midfielder')

sns.countplot(x='foot', data=df_A, ax=axes[1, 0])
axes[1, 0].set_title('Attacker')

sns.countplot(x='foot', data=df_D, ax=axes[1, 1])
axes[1, 1].set_title('Defender')

plt.tight_layout()
plt.show()
```
